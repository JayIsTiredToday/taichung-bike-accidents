```python
import pandas as pd 
import glob 
df = pd.DataFrame()
src = 'incident_data/'
for file in glob.glob(src + '*.csv'):
    temp = pd.read_csv(file)
    temp = temp[(temp['當事者區分(類別)'] == 'F01') | (temp['當事者區分(類別)'] == 'F02')]
    df = pd.concat([df, temp], axis=0) 
```

```python
df.to_csv('incidents_clean.csv')
```
